```python
def data_type(data):
    int = 0
    float = 0
    object = 0
    other = 0
    
    for i in data.columns:
        print(f"feature_name {i}:{data[i].dtype}")
        if data[i].dtype == "int32" or data[i].dtype == "int64":
            int = int+1 
        elif data[i].dtype == "float32" or data[i].dtype == "float64":
            float = float+1 
        elif data[i].dtype == "object":
            object = object+1
        
        else:
            other = other+1
            
    print("Data summary")
    print("_------------")
    print("No of integers ",int)
    print("No of  float features",float)
    print('No of object',object)
    print('No of other',other)
        
        
```

```python
data_type(df)
```

```python
# if any feature in a data set has more than 10 (or) 15 unique categories we refer it as a cardinal data and the problem is termed as 'handaling high' cardinality 
```

```python
# implementing the first variant (count frequency encoder)
# steps:
# 1. Identify the top  categories with highest occurances
# 2. APPLY one hot encoding on these n categories
#Not: The value of n is sbjective and can be changed to get better results usually value of n is choosen in such a way that it constitute to 80% of the data
```

```python
df_copy = df
```

```python
def count_frequency_encoder(data):
    for i in data.columns:
        if data[i].dtype == "object":
            top_n = data[i].value_counts().head().index
            for j in top_n:
                data[i+'_'+j] = np.where(data[i] == j,1,0)
            data.drop(i,axis=1,inplace=True)
    return data
            
            
    
            
   
            
    
```

```python
count_frequency_encoder(df)
```

```python
def encoding_features(ds,n):
    for i in ds.columns:
        if ds[i].dtype == 'object':
            top_n = ds[i].value_counts().head(int(len(ds[i]*(n/100)))).index
            for j in top_n:
                ds[i+'_'+j] = np.where(ds[i] == j,1,0)
            ds.drop(i,axis=1,inplace=True)
    return ds
        
       
```

```python
encoding_features(df_copy,80)
```
